```python
import matplotlib.pyplot as plt
import numpy as np

size = 10
grid = 12

img = np.zeros((grid*size, grid*size))

for i in range(grid):
    for j in range(grid):
        if (i + j) % 2 == 0:
            shift = int(size * 0.3)
        else:
            shift = -int(size * 0.3)

        x = i * size
        y = j * size

        img[x:x+size, y+shift:y+size+shift] = 1

plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt

rows, cols = 12, 12
block = 20

fig, ax = plt.subplots()

for i in range(rows):
    for j in range(cols):
        x = j * block
        y = i * block

        if i % 2 == 0:
            x += block // 2  # offset

        color = 'black' if (i + j) % 2 == 0 else 'white'

        rect = plt.Rectangle((x, y), block, block, color=color)
        ax.add_patch(rect)

ax.set_xlim(0, cols*block + block)
ax.set_ylim(0, rows*block)
ax.set_aspect('equal')
ax.axis('off')

plt.gca().invert_yaxis()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

for r in range(20, 200, 20):
    theta = np.linspace(0, 2*np.pi, 200)

    for i in range(len(theta)-1):
        if i % 2 == 0:
            color = 'black'
        else:
            color = 'white'

        x = r * np.cos(theta[i:i+2])
        y = r * np.sin(theta[i:i+2])

        ax.plot(x, y, color=color, linewidth=6)

ax.set_aspect('equal')
ax.axis('off')
plt.show()
```
