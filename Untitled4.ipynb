```python
from ColabTurtle.Turtle import *

initializeTurtle()
bgcolor("black")
speed(13)   # fastest

colors = ["yellow", "red", "pink", "cyan", "lightgreen", "blue"]

# ⚡ Optimized arc (fewer steps, larger angle)
def fast_arc(radius, angle):
    steps = 12   # reduced steps (big speed gain)
    step_angle = angle / steps
    step_length = (2 * 3.1416 * radius) * (step_angle / 360)

    for _ in range(steps):
        forward(step_length)
        left(step_angle)

# ⚡ Main drawing
for i in range(120):   # reduced iterations (faster but still beautiful)
    pencolor(colors[i % 6])

    fast_arc(190 - i/2, 90)
    left(90)

    fast_arc(190 - i/3, 90)
    left(60)

hideturtle()
```

```python
import math
from ColabTurtle.Turtle import *

initializeTurtle()
bgcolor("black")
speed(13)

colors = ["cyan", "magenta", "yellow", "white"]

# Optimized circle function for speed
def custom_circle(radius, extent=360):
    if extent == 0:
        return

    # REDUCED STEPS: Instead of 1 step per degree, we use 10 steps total for the arc
    # This makes it roughly 6x faster for a 60-degree arc
    steps = 10
    angle_per_step = float(extent) / steps

    # Calculate segment length based on arc circumference logic
    # Formula: (2 * pi * r) * (angle_per_step / 360)
    segment_length = (2 * math.pi * abs(radius) * abs(angle_per_step)) / 360
    turn_direction_func = left if radius >= 0 else right

    for _ in range(steps):
        forward(segment_length)
        turn_direction_func(angle_per_step)

def petal():
    for _ in range(2):
        custom_circle(100, 60)
        left(120)

# Main drawing loop
for i in range(36):
    pencolor(colors[i % 4])
    petal()
    left(10)

hideturtle()
```

```python
from ColabTurtle.Turtle import *

initializeTurtle()
bgcolor("black")
speed(13)

colors = ["cyan","blue","white"]

def fast_arc(radius, angle):
    steps = 10
    step_angle = angle / steps
    step_length = (2 * 3.1416 * radius) * (step_angle / 360)

    for _ in range(steps):
        forward(step_length)
        left(step_angle)

for i in range(100):
    pencolor(colors[i % 3])
    fast_arc(150 - i, 120)
    left(137)

hideturtle()
```
